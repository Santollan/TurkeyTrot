```python
import requests
import json
import csv
import os

def get_race_results(url):
    response = requests.get(url)

    if response.status_code == 200:
        json_content = response.json()

        if 'data' in json_content:
            return json_content['data']
        else:
            print(f"No race results found in the JSON data for URL: {url}")
            return []
    else:
        print(f"Failed to retrieve the page. Status code: {response.status_code}")
        return []

# List of URLs to fetch data from
urls_5k = [
    'https://results.raceroster.com/v2/api/result-events/64539/sub-events/187093/results?start=0&limit=1000&locale=en-US',
    'https://results.raceroster.com/v2/api/result-events/64539/sub-events/187093/results?start=1000&limit=1000&locale=en-US',
    'https://results.raceroster.com/v2/api/result-events/64539/sub-events/187093/results?start=2000&limit=1000&locale=en-US',
    'https://results.raceroster.com/v2/api/result-events/64539/sub-events/187093/results?start=3000&limit=1000&locale=en-US',
    'https://results.raceroster.com/v2/api/result-events/64539/sub-events/187093/results?start=4000&limit=1000&locale=en-US',
    'https://results.raceroster.com/v2/api/result-events/64539/sub-events/187093/results?start=5000&limit=1000&locale=en-US',
    'https://results.raceroster.com/v2/api/result-events/64539/sub-events/187093/results?start=6000&limit=1000&locale=en-US'
]

urls_8k = [
    'https://results.raceroster.com/v2/api/result-events/64539/sub-events/187092/results?start=0&limit=1000&locale=en-US&uuid=38fb69f1-e9ec-4c29-8cdf-e82184cc08fe',
    'https://results.raceroster.com/v2/api/result-events/64539/sub-events/187092/results?start=1000&limit=1000&locale=en-US&uuid=38fb69f1-e9ec-4c29-8cdf-e82184cc08fe',
    'https://results.raceroster.com/v2/api/result-events/64539/sub-events/187092/results?start=2000&limit=1000&locale=en-US&uuid=38fb69f1-e9ec-4c29-8cdf-e82184cc08fe'
]

desktop_path = os.path.join(os.path.join(os.path.expanduser('~')), 'Desktop')
csv_file_path = os.path.join(desktop_path, 'race_results_combined2.csv')

# Open the CSV file in write mode with UTF-8 encoding
with open(csv_file_path, 'w', newline='', encoding='utf-8') as csvfile:
    # Create a CSV writer object
    csv_writer = csv.writer(csvfile)

    # Iterate through each URL for 5k race results and get race results
    for url in urls_5k:
        race_results = get_race_results(url)

        if race_results:
            # Write the header to the CSV file for 5k race
            csv_writer.writerow(["Place", "Bib", "Name", "Chip Time", "Gun Time", "Pace", "Gender", "Age", "Race Type"])
            
            # Iterate through each result and write to the CSV file
            for result in race_results:
                place = result.get('overallPlace', '')
                bib = result.get('bib', '')
                name = result.get('name', '')
                chip_time = result.get('chipTime', '')
                gun_time = result.get('gunTime', '')
                pace = result.get('overallPace', '')
                gender = result.get('genderSexId', '')
                age = result.get('age', '')

                # Write the result information to the CSV file for 5k race
                csv_writer.writerow([place, bib, name, chip_time, gun_time, pace, gender, age, "5k"])

    # Iterate through each URL for 8k race results and get race results
    for url in urls_8k:
        race_results = get_race_results(url)

        if race_results:
            # Write the header to the CSV file for 8k race
            csv_writer.writerow(["Place", "Bib", "Name", "Chip Time", "Gun Time", "Pace", "Gender", "Age", "Race Type"])
            
            # Iterate through each result and write to the CSV file
            for result in race_results:
                place = result.get('overallPlace', '')
                bib = result.get('bib', '')
                name = result.get('name', '')
                chip_time = result.get('chipTime', '')
                gun_time = result.get('gunTime', '')
                pace = result.get('overallPace', '')
                gender = result.get('genderSexId', '')
                age = result.get('age', '')

                # Write the result information to the CSV file for 8k race
                csv_writer.writerow([place, bib, name, chip_time, gun_time, pace, gender, age, "8k"])

print(f"Race results have been saved to {csv_file_path}")
```
